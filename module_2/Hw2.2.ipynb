Удалите все столбцы с пропусками, кроме столбца "Язык.программирования"

```python
new_list = []
for i,j in enumerate(pandas.isna(dataset).sum().index):
    if pandas.isna(dataset).sum()[i] != 0 and j != "Язык.программирования":
        new_list.append(j)
dataset = dataset.drop(new_list, axis=1)
```

Создайте новую таблицу python_data в которой будут только строки со специалистами указавшими язык программирования Python

Используя метод groupby выполните группировку по столбцу "Должность"

```python
group_python = python_data.groupby(["Должность"]).aggregate(sum)
```

Создайте новый DataFrame, где для сгруппированных данных по столбцу "Должность", выполните агрегацию данных с помощью метода agg и найдите минимальное и максимальное значение в столбце "Зарплата.в.месяц"

```python
group_python.agg({"Зарплата.в.месяц" : ['max', 'min']})
```

Создайте функцию fill_avg_salary которая будет возвращать среднее значение заработной платы в месяц. Используйте ее для метода apply и создайте новый столбик "avg"

```python
len(python_data[python_data["Должность"] == "Junior Software Engineer"])
list_of_jobs = ["Junior Software Engineer", "Senior Software Engineer", "Software Engineer", "System Architect", "Technical Lead"]
new_list = []
for i,j in enumerate(list_of_jobs):
    new_list.append(round(group_python["Зарплата.в.месяц"][i]/len(python_data[python_data["Должность"] == j])))
    
group_python["avg"] = new_list
```

Создайте описательную статистику с помощью метода describe для нового столбца.

```python
group_python["avg"].describe = "In this avg column you can see detailed information about average income/salary on this position!"
```

Сохраните полученную таблицу в CSV файл

```python
group_python.to_csv("group_python.csv", index=True)
```
