```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from scipy import integrate

speed = [25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65]
```

```python
time = np.linspace(0, 11, num=12)
print(time)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
plt.scatter(time, speed)
plt.plot(time, speed, '--b')
plt.ylim(0, 130)
plt.xlim(0, 12)
plt.grid ( True )
plt.show()
```

```python
f = interpolate.interp1d(time, speed, kind='cubic') 

plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, num=1000)
plt.plot(more_date, f(more_date))
```

```python
result = integrate.quad(f,0,11)
```

```python
f = interpolate.interp1d(time, speed, kind='quadratic') 

plt.plot(time, speed, 'o')
more_date = np.linspace(0, 11, num=1000)
plt.plot(more_date, f(more_date))
```

```python
result = integrate.quad(f,0,11)
```
